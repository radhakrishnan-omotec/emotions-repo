```python
# Define the path to the input file
input_file_path = r'/content/emotions-repo/Prototype-1/gsr_data.csv'
output_file_path = r'/content/emotions-repo/depression_stage_data.csv'
```

### Read the input CSV file

```python
# Read the input CSV file
data = pd.read_csv(input_file_path)
```

```python
data.head()
```

### Create a new DataFrame with the required columns

```python
# Create a new DataFrame with the required columns
output_data = pd.DataFrame()
```

### Copy the 'Heart Rate (bpm)' column from the input CSV

```python
# Copy the 'Heart Rate (bpm)' column from the input CSV
output_data['Heart Rate (bpm)'] = data['GSR Value']
```

```python
output_data.head()
```

### Add the 'SpO2 (%)' column with a constant value of 85


```python
# Add the 'SpO2 (%)' column with a constant value of 85
output_data['SpO2 (%)'] = 85
```

```python
output_data.head()
```

### Add the 'Depression Stage' column based on conditions


```python
# Add the 'Depression Stage' column based on conditions
output_data['Depression Stage'] = output_data['Heart Rate (bpm)'].apply(
    lambda x: 0 if x < 2000 else (1 if 2001 <= x <= 2200 else (2 if 2201 <= x <= 2300 else 3))
)
```

```python
output_data.head(50)
```

### Save the new DataFrame to a CSV file


```python
# Save the new DataFrame to a CSV file
output_data.to_csv(output_file_path, index=False)
```
